```python
%%time
%load_ext lab_black
%config Completer.use_jedi = False

import pandas as pd
from tqdm import tqdm
import numpy as np
from utils import dealing_results

# from collections import defaultdict
ALL_METHODS=[]
FILES = [
    "./RESULT_EXCELS/Decreasing_Data.csv",
    "./RESULT_EXCELS/Increasing_Data.csv",
    "./RESULT_EXCELS/Mild_Fluctuate_Data.csv",
    "./RESULT_EXCELS/All_Data.csv",
]
```

```python
%%time
# Arithmetic
from Arithmetic import speedy_grid_trading
from utils import dealing_results

SD_PARAMETER_VALUE = 3
INTERVAL_FOR_BOUNDS_VALUE = 5
BEGINNING_CASH = 100000
METHOD_TAG = "Arithmetic"
for FILENAME in FILES:
    data = pd.read_csv(FILENAME)
    print(f"\nData: {FILENAME}")
    records_df, final_profit = speedy_grid_trading(
        data,
        INTERVAL_FOR_BOUNDS=INTERVAL_FOR_BOUNDS_VALUE * 24 * 60,
        SD_PARAMETER=SD_PARAMETER_VALUE,
    )
    records_df["Hour"] = pd.to_datetime(records_df["Hour"])
    dumb_profit, difference = dealing_results(
        records_df=records_df,
        FILENAME=FILENAME,
        BEGINNING_CASH=BEGINNING_CASH,
        final_profit=final_profit,
        data=data,
        METHOD_TAG=METHOD_TAG,
    )
    print(f"dumb_profit:{dumb_profit}, difference:{difference}")
    ALL_METHODS.append([METHOD_TAG, FILENAME, final_profit, dumb_profit, difference])
```
