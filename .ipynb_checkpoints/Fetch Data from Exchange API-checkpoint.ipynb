```python
import ccxt
```

```python
import time
```

```python
exchange = ccxt.huobipro()
```

```python
#Ticker data
#data = exchange.fetchTicker(symbol = 'BTC/USDT')
```

```python
# K graph data
data = exchange.fetch_ohlcv(symbol = 'BTC/USDT', timeframe = '1h', limit = 2000)
```

```python
print(data)
```

```python
import pandas as pd
df = pd.concat([pd.Series(x) for x in data], axis=1)
```

```python
df
```

```python
df = df.T
```

```python
df.columns = ['TimeStamp', 'Open', 'High', 'Low', 'Close', 'Volume'] 
```

```python
df
```

```python
df.to_csv('Huobi_BTC_1h.csv')
```

```python
from datetime import datetime
df['TimeStamp'] = pd.to_datetime(df['TimeStamp'])
```
