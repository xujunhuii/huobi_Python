```python
%load_ext lab_black
%config Completer.use_jedi = False
import pandas as pd
from tqdm import tqdm
import time
import numpy as np
```

```python
def calc_ATRP(high, low, close, n=15 * 24 * 60):
    high_low = high - low
    high_cp = np.abs(high - close.shift())
    low_cp = np.abs(low - close.shift())

    df = pd.concat([high_low, high_cp, low_cp], axis=1)
    true_range = np.max(df, axis=1)
    average_true_range = true_range.rolling(n).mean()
    average_true_range_percentage = average_true_range / close

    return average_true_range_percentage


##显示ATRP数值
FILENAME = "Binance_BTC_1m.csv"
data = pd.read_csv(FILENAME)
ATRP = calc_ATRP(data["High"], data["Low"], data["Close"])
data["ATRP"] = ATRP
data
```

```python
TOTAL_RANGE = data["ATRP"].max() - data["ATRP"].min()
AVERAGE = (data["ATRP"].max() + data["ATRP"].min()) / 2
data["ATRP_Diff"] = (data["ATRP"] - AVERAGE) / TOTAL_RANGE
data["ATRP_Diff"].min()
data
```
