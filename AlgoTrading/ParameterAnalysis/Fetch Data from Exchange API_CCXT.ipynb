```python
%load_ext lab_black
import ccxt
import time
import pandas as pd
```

```python
exchange = ccxt.binance()
```

```python
# Ticker data
# data = exchange.fetchTicker(symbol = 'BTC/USDT')
```

```python
current_time = int(time.time() // 60 * 60 * 1000)  # ms
print(current_time)
limit = 100
```

```python
%%time
data_list = []
for i in range(1,2):
#     data = exchange.fetch_ohlcv(symbol = 'BTC/USDT', timeframe = '1m', limit = 100, since = current_time - limit*60*1000*i) #min
    data = exchange.fetch_ohlcv(symbol = 'BTC/USDT', timeframe = '1h', limit = 100, since = current_time - limit*60*60*1000*i) #hour
    data_list.append(data)
    #df = pd.DataFrame(data)
    #print(data)
data_list
```

```python
len(data_list)
```

```python
concatenated = []
for l in data_list:
    concatenated.extend(l)
```

```python
# concatenated
```

```python
df = pd.DataFrame(concatenated)
# df
```

```python
df.columns = ["TimeStamp", "Open", "High", "Low", "Close", "Volume"]
# df
```

```python
df = df.sort_values(by=["TimeStamp"], ascending=True)
df
```

```python
df.dtypes
```

```python
df["TimeStamp"] = pd.to_datetime(df["TimeStamp"], unit="ms") + pd.Timedelta(hours=8)
```

```python
df = df.set_index("TimeStamp", drop="True")
```

```python
pd.set_option("display.max_rows", 500)
print(df.isnull().values.any())
df
```
