```python
#%load_ext lab_black
#%config Completer.use_jedi = False
import pandas as pd
from Grid_Tuning import grid_trading, speedy_grid_trading
from tqdm import tqdm
import time
```

```python
%%time

FILENAME = "Binance_BTC_1m.csv"
data = pd.read_csv(FILENAME)
optimization=[]
for interval_for_bounds in tqdm(range(5,20)):
    for sd in range(2,5):
        records_df, final_profit = speedy_grid_trading(
            data, INTERVAL_FOR_BOUNDS=interval_for_bounds * 24 * 60, SD_PARAMETER=sd,
        )
        optimization.append([final_profit, "", interval_for_bounds, sd])
```
