```python
%load_ext lab_black
%config Completer.use_jedi = False
import pandas as pd

f = "Binance_BTC_1m.csv"
data = pd.read_csv(f)
data
N = 24 * 60
M = 2
CLOSE = "Close"
TIMESTAMP = "TimeStamp"


def calc_RSV(prices, n=9):
    C = prices
    H = prices.rolling(n).max()
    L = prices.rolling(n).min()
    RSV = (C - L) / (H - L)
    return RSV


RSV = calc_RSV(data["Close"])
data["RSV"] = RSV
```

```python
def calc_bbands(prices, n=N, m=M):
    middle_vals = prices.rolling(n).mean()
    std_vals = prices.rolling(n).std(ddof=0)
    upper_vals = middle_vals + m * std_vals
    lower_vals = middle_vals - m * std_vals
    return upper_vals, middle_vals, lower_vals


upper, middle, lower = calc_bbands(data["Close"])
data["boll_upper"] = upper
# data['boll_middle'] = middle
data["boll_lower"] = lower
data = data.dropna()
data.reset_index(drop=True, inplace=True)
data
```

```python
BEGINNING_CASH = 100000


def buying(
    last_price, trading_buying_price, cash, n, r, records, price, time, index,
):
    n += cash // trading_buying_price
    cash -= trading_buying_price * n
    profit = cash - BEGINNING_CASH
    records.append(["Buying", -round(price), n, round(cash), round(profit), time])
    last_price = price
    return last_price, cash, n, profit, records, index + 1
```

```python
def selling(last_price, trading_selling_price, cash, n, r, records, price, time, index):
    cash += trading_selling_price * n
    profit = cash - BEGINNING_CASH
    records.append(["Selling", round(price), 0, round(cash), round(profit), time])
    last_price = price
    n = 0
    start_index = index + 1
    upper_bound = row["boll_upper"]
    lower_bound = row["boll_lower"]
    return last_price, cash, n, profit, records, start_index
```

```python
def transact(price, last_price, cash, n, profit, records, time, index):
    trading_buying_price = price * (1 + FEE)
    trading_selling_price = price * (1 - FEE)
    if last_price == 0 or (
        n == 0
        and trading_buying_price <= last_price * (1 - r)
        and cash - trading_buying_price > 0
    ):
        last_price, cash, n, profit, records, index = buying(
            last_price, trading_buying_price, cash, n, r, records, price, time, index,
        )
    elif n > 0 and trading_selling_price >= last_price * (1 + r):
        last_price, cash, n, profit, records, index = selling(
            last_price, trading_selling_price, cash, n, r, records, price, time, index
        )
    return last_price, cash, n, profit, records, index
```

```python
FEE = 0.005
records = []
profit = 0
cash = BEGINNING_CASH
n = 0
last_price = 0
upper_bound = data.loc[0]["boll_upper"]
lower_bound = data.loc[0]["boll_lower"]
start_index = 0
r = 0.005


def get_profit_by_RSV(last_price, cash, n, profit, records, start_index):
    for index, row in data[start_index:].iterrows():
        time = row[TIMESTAMP]
        price = row[CLOSE]
        if lower_bound <= price <= upper_bound:
            if last_price < lower_bound:
                last_price = 0
            last_price, cash, n, profit, records,start_index = transact(
                price, last_price, cash, n, profit, records, time,index
            )
            break
    return last_price, cash, n, profit, records, start_index


while start_index < len(data):
    last_price, cash, n, profit, records, start_index = get_profit_by_RSV(
        last_price, cash, n, profit, records, start_index
    )
    print(profit, cash, start_index)
```
