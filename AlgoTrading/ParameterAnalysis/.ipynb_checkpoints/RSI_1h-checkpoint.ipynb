```python
%load_ext lab_black
import pandas as pd
```

```python
import json

f = open("histohour.json",)

data = json.load(f)
data = pd.DataFrame.from_dict(data["Data"])
data = data[["time", "close"]]
f.close()
data
```

```python
def get_profit(lower_limit, upper_limit):
    i = 0
    profit = 0
    records = []
    n = 0
    for index, row in data.iterrows():
        day = index
        price = row["close"]
        if i % 2 == 0:
            if row["RSI"] < lower_limit and n == 0:
                profit -= price
                records.append(["Buying", -round(price), round(profit), day])
                n += 1
                i += 1
        else:
            if row["RSI"] > upper_limit and n > 0 and profit + price > 0:
                profit += price
                records.append(["Selling", round(price), round(profit), day])
                n -= 1
                i += 1
    return profit


window_length = 14

delta = data["close"].diff()
# delta = delta[1:]
up, down = delta.copy(), delta.copy()
up[up < 0] = 0
down[down > 0] = 0
roll_up1 = up.ewm(span=window_length).mean()
roll_down1 = down.abs().ewm(span=window_length).mean()
RS1 = roll_up1 / roll_down1
RSI1 = 100.0 - (100.0 / (1.0 + RS1))
data.loc[:, "RSI"] = RSI1
data.loc[:, "TimeStamp"] = pd.to_datetime(data["time"], unit="s")
data = data.set_index("TimeStamp")
get_profit(50, 60)
```

```python
%time
import datetime
from tqdm import tqdm

i = 0
profit = 0
records = []
n = 0
for index, row in tqdm(data.iterrows()):
    #     day = datetime.datetime.strptime(index, "%Y-%m-%d %H:%M:%S")
    day = index
    price = row["close"]
    if i % 2 == 0:
        if row["RSI"] < 50 and n == 0:
            profit -= price
            print(
                f"Buying 1 BTC @ USD {round(price)}, total profit: {round(profit)} on {day}"
            )
            records.append(["Buying", -round(price), round(profit), day])
            n += 1
            i += 1
    else:
        if row["RSI"] > 60 and n > 0 and profit + price > 0:
            profit += price
            print(
                f"Selling 1 BTC @ USD {round(price)}, total profit: {round(profit)} on {day}"
            )
            records.append(["Selling", round(price), round(profit), day])
            n -= 1
            i += 1
print(f"\nTotal profit: {round(profit)}")
# df.loc[start_time:]
# records
```

```python
formated_profit = f"{round(profit):,}"
print(f"Total profit: USD {formated_profit}\n")
records_df = pd.DataFrame(records, columns=["Action", "Price", "Profit", "Day"])
pd.set_option("max_rows", 30)
records_df
```
